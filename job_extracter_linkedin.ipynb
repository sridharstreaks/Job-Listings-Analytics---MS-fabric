```python
def random_sleep_time():
    # prints a random value from the list
    sleep_time = [3,4,5,6,7,8,9,10]
    rand=random.choice(sleep_time)
    return rand
```

```python
def job_id_collector(job_pages,job_titles,location):
    for i in range(0, job_pages):
        product = start * i * 25
        print(product)
        list_url = f"https://www.linkedin.com/jobs-guest/jobs/api/seeMoreJobPostings/search?keywords={job_titles}&location={location}&start={start}&f_TPR=r86400"
        response = requests.get(list_url)
        if response.status_code==200:

            list_data = response.text
            list_soup = BeautifulSoup(list_data, "html.parser")
            page_jobs = list_soup.find_all("li")

            for job in page_jobs:
                base_card_div = job.find("div", {"class": "base-card"})
                job_id_num = base_card_div.get("data-entity-urn").split(":")[3]
                if job_id_num!=None and job_id_num not in id_list:
                    id_list.append(job_id_num)
                else:
                    continue
            time.sleep(random_sleep_time())
        else:
            continue
```

```python
id_list = []
job_titles = "data analyst"  
location = "india"  
start = 1

job_id_collector(job_pages,job_titles,location)
    
time.sleep(random_sleep_time())

for job_id_num in id_list:
    job_url = f"https://www.linkedin.com/jobs-guest/jobs/api/jobPosting/{job_id_num}"

    job_response = requests.get(job_url)
    if job_response.status_code==200:
        job_soup = BeautifulSoup(job_response.text, "html.parser")

        job_id.append(job_id_num)

        temp_job_title=job_soup.find("h2", {"class":"top-card-layout__title font-sans text-lg papabear:text-xl font-bold leading-open text-color-text mb-0 topcard__title"}).text.strip().lower()

        if temp_job_title==job_titles or job_titles in temp_job_title:
            try:
                job_title.append(job_soup.find("h2", {"class":"top-card-layout__title font-sans text-lg papabear:text-xl font-bold leading-open text-color-text mb-0 topcard__title"}).text.strip())
            except:
                job_title.append(None)

            try:
                company_name.append(job_soup.find("a", {"class": "topcard__org-name-link topcard__flavor--black-link"}).text.strip())
            except:
                company_name.append(None)

            try:
                job_Description.append(job_soup.find("div",{"class":"description__text description__text--rich"}).text.strip())
            except:
                job_Description(None)

            try:
                time_posted.append(job_soup.find("span", {"class": "posted-time-ago__text topcard__flavor--metadata"}).text.strip())
            except:
                time_posted.append(job_soup.find("span",{"class": "posted-time-ago__text posted-time-ago__text--new topcard__flavor--metadata"}).text.strip())

            if job_soup.find("span", {"class": "num-applicants__caption topcard__flavor--metadata topcard__flavor--bullet"}) is not None:
                temp_num_applicants=job_soup.find("span", {"class": "num-applicants__caption topcard__flavor--metadata topcard__flavor--bullet"}).text.strip()
            elif job_soup.find("figcaption", {"class": "num-applicants__caption"}) is not None:
                temp_num_applicants=job_soup.find("figcaption", {"class": "num-applicants__caption"}).text.strip()

            num_applicants.append(int(''.join(filter(str.isdigit, temp_num_applicants))))

            span_elements=job_soup.find_all("span", {"class":"description__job-criteria-text description__job-criteria-text--criteria"})
            # Extract text from each span element
            for i in range(len(variables)):
                if span_elements[i].get_text(strip=True) is not None:
                    variables[i].append(span_elements[i].get_text(strip=True))
                else:
                    continue

            time.sleep(random_sleep_time())

        elif temp_job_title==None:
            continue
        else:
            continue
    else:
        id_list.remove(job_id_num)
        time.sleep(random_sleep_time())
        continue
```
